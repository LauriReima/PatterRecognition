Import all the packages needed for this notebook in one cell (add the ones you apply):

```python
import pandas as pd
import numpy as np
from random import sample
from random import seed
import glob, os
import cv2 as cv
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis
import itertools
```

```python
def calculate_features(contour, ellipse):
    # Major axis length of the ellipse
    major_axis_length = max(ellipse[1])

    # Minor axis length of the ellipse
    minor_axis_length = min(ellipse[1])

    # Area inside the contour
    area = cv2.contourArea(contour)

    # Perimeter of the contour
    perimeter = cv2.arcLength(contour, closed=True)

    # Equivalent diameter
    equivalent_diameter = np.sqrt(4 * area / np.pi)

    # Compactness
    compactness = perimeter**2 / (4 * np.pi * area)

    # Shape Factor 1 (SF1)
    sf1 = (perimeter**2) / area

    # Shape Factor 2 (SF2)
    sf2 = 4 * np.pi * area / (perimeter**2)

    return major_axis_length, minor_axis_length, area, perimeter, equivalent_diameter, compactness, sf1, sf2
```

Be aware that *drawContours* adds the contour to the original image, so use the copy of the test_image as input for the function!

```python
# https://docs.opencv.org/4.x/d4/d73/tutorial_py_contours_begin.html
def find_contour(im):
    imgray = cv.cvtColor(im, cv.COLOR_BGR2GRAY)
    ret, thresh = cv.threshold(imgray, 127, 255, 0)
    contours, hierarchy = cv.findContours(thresh,
                                          cv.RETR_TREE,
                                          cv.CHAIN_APPROX_SIMPLE)
    cnt = contours
    
    return(cnt.reshape(-1,2)) 
```

## Feature extraction

Gather the color feature data (12 features): <br>

First, convert each RGB image to YCbCr-image. You can use *cvtColor* function from OpenCV with appropriate color conversion. <br>
Be aware that openCV default channel order is BGR not RGB. In which order does the *cvtColor* function output the components? <br>

Calculate the following color features for each image, including only the pixels within the contour (you can use *pointPolygonTest* from OpenCV)
- Mean for each YCbCr component
- Variance for each YCbCr component
- Skewness for each YCbCr component
- Kurtosis for each YCbCr component

```python
 # Function to calculate mean, variance, skewness, and kurtosis
columns = ['Image_Path', 'Major_Axis_Length', 'Minor_Axis_Length', 'Area', 'Perimeter', 'Equivalent_Diameter', 'Compactness', 'SF1', 'SF2']

def calculate_statistics(data):
    mean_value = np.mean(data)
    variance_value = np.var(data)
    skewness_value = skew(data, axis=None)
    kurtosis_value = kurtosis(data, axis=None)
    return mean_value, variance_value, skewness_value, kurtosis_value

subset_data_list = []
# Iterate over each image in the subset
for image_path in subset:
    image = cv2.imread(image_path)
    
    # Convert RGB image to YCbCr
    ycbcr_image = cv2.cvtColor(image, cv2.COLOR_BGR2YCrCb)
    gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    _,thresh_test = cv2.threshold(gray_img, 127,255,cv2.THRESH_BINARY)
    # Get the contour for the current image
    contours, _ = cv2.findContours(thresh_test, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    ellipse = cv2.fitEllipse(contours[0])
    #calculate_features(contour, ellipse)
    # Iterate over the contours
    for contour in contours:
        # Create a mask for the current contour
        mask = np.zeros_like(ycbcr_image[:, :, 0], dtype=np.uint8)
        cv2.drawContours(mask, [contour], -1, (255, 255, 255), thickness=cv2.FILLED)

         # Use cv2.bitwise_and to include only pixels within the contour
        ycbcr_contour = cv2.bitwise_and(ycbcr_image, ycbcr_image, mask=mask)
        
        # Extract Y, Cb, Cr components
        y_component, cb_component, cr_component = cv2.split(ycbcr_contour)

        # Print the shape of the Y component and the mask
        #print("Y Component Shape:", y_component.shape)
        #print("Mask Shape:", mask.shape)
        
        # Calculate statistics for each component
        y_stats = calculate_statistics(y_component)
        cb_stats = calculate_statistics(cb_component)
        cr_stats = calculate_statistics(cr_component)
        point = cv2.pointPolygonTest(contour, (125, 160), False)
        

        subset_data_list.append({
            'Image_Path': image_path,
            'Y_Mean': y_stats[0], 'Y_Variance': y_stats[1], 'Y_Skewness': y_stats[2], 'Y_Kurtosis': y_stats[3],
            'Cb_Mean': cb_stats[0], 'Cb_Variance': cb_stats[1], 'Cb_Skewness': cb_stats[2], 'Cb_Kurtosis': cb_stats[3],
            'Cr_Mean': cr_stats[0], 'Cr_Variance': cr_stats[1], 'Cr_Skewness': cr_stats[2], 'Cr_Kurtosis': cr_stats[3],
   
        })
```

Does a high value of compactness mean that the rice is rather round or thin?

Compactness is a measure that describes how compact or round an object is. Specifically, it is defined as the ratio of the object's area to the area of a circle with the same perimeter. The formula for compactness is:

Compactness = Area / Perimeter**2

A high value of compactness indicates that the object is more compact or round. In the context of rice grains, a high compactness value would suggest that the rice grain is relatively round or compact. Conversely, a low compactness value would suggest that the rice grain is more elongated or thin.

A high value of compactness generally means that the rice is more round, not thin.
